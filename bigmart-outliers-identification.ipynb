# using boxplot for finding outliers 
## First Finding the Univariate Outliers


```python
# Looking for Univariate Outliers

plt.rcParams['figure.figsize'] = (16,4)

plt.subplot(1,4,1)
sbn.boxplot(data['Item_Weight'])

plt.subplot(1,4,2)
sbn.boxplot(data['Item_Visibility'])

plt.subplot(1,4,3)
sbn.boxplot(data['Item_MRP'])

plt.subplot(1,4,4)
sbn.boxplot(data['Item_Outlet_Sales'])

plt.suptitle("Univariate Outliers")
plt.show()
```

## Finding the bivariate outliers

Finding bivariate outliers. We usually compare the variables with target.

```python
plt.subplot(1,3,1)
sbn.scatterplot(y=data['Item_Outlet_Sales'], x= data['Item_Weight'])

plt.subplot(1,3,2)
sbn.scatterplot(y=data['Item_Outlet_Sales'], x= data['Item_Visibility'])

plt.subplot(1,3,3)
sbn.scatterplot(y=data['Item_Outlet_Sales'], x= data['Item_MRP'])
```
